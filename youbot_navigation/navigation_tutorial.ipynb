## 1. Launch the YouBot Description
Start by launching the description package. This will bring up the YouBot in Gazebo and RViz, with the joint state publisher and the xterm teleop twist menu.

```python
# Launch the YouBot description in simulation
ros2 launch youbot_navigation.launch.py
```

## 2. Run SLAM (Simultaneous Localization and Mapping)
To perform SLAM, use the following command. This will start the SLAM toolbox with the specified parameters file. Make sure to use the correct path to your SLAM parameters YAML file.

```python
# Run SLAM with the specified parameters
ros2 launch youbot_navigation slam.launch.py slam_params_file:=/home/youssef/fyp_ws/src/youbot_navigation/config/slam_toolbox_params.yaml use_sim_time:=true
```

## 3. Save and Serialize the Map
After mapping, save and serialize your map. Then, update the `mapper_params_online_async.yaml` file to include the path to your serialized map (with `.serial` or `.yaml` extension).

## 4. Launch Localization (Online Async)
Launch the SLAM toolbox in online async mode for localization, using the updated parameters file.

```python
# Launch localization with the serialized map
ros2 launch slam_toolbox online_async_launch.py slam_params_file:=/home/youssef/fyp_ws/src/youbot_navigation/config/mapper_params_online_async.yaml
```

**Tip:** In RViz, set the fixed frame to `/map` to visualize the loaded map.

## 5. Launch Navigation2 Stack
While localization is running, launch the Navigation2 stack. This will enable navigation and path planning.

```python
# Launch the Navigation2 stack
ros2 launch youbot_navigation nav2.launch.py
```

In RViz, select the `/global_cost_map` and use the `goal_pose` tool to send navigation goals to the robot.
